```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('awards.csv')

# Convert the 'date_employed' column to datetime using 'month/day/year' format
df['date_employed'] = pd.to_datetime(df['date_employed'], format='%m/%d/%Y')

# Define the target years
target_years = [2024, 2025, 2026]

# Create an empty dictionary to store DataFrames for each year
yearly_data = {}

# Loop through the target years
for target_year in target_years:
    # Calculate the number of years spent in the company for the specific year
    df['years_spent'] = target_year - df['date_employed'].dt.year
    
    # Create a result DataFrame for the specific year
    result_df = pd.DataFrame({'Company': df['company'], 'Full Name': df['full_name'], 'Years Spent': df['years_spent']})
    
    # Filter the data for the specific year
    yearly_data[target_year] = result_df.copy()

# Create an Excel writer object to save multiple DataFrames into a single Excel file
with pd.ExcelWriter('awards_years.xlsx', engine='xlsxwriter') as writer:
    for year, data in yearly_data.items():
        # Save each DataFrame to a separate sheet with the year as the sheet name
        data.to_excel(writer, sheet_name=str(year), index=False)

print("Data organized and saved to 'awards_years.xlsx'")
```

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('awards.csv')

# Convert the 'date_employed' column to datetime using 'month/day/year' format
df['date_employed'] = pd.to_datetime(df['date_employed'], format='%m/%d/%Y')

# Define the years you want to check
target_years = [2024, 2025, 2026]

# Create a result list to store the results
results = []

# Loop through the target years and collect the names of people who have spent exactly those years
for year in target_years:
    # Calculate the number of years spent in the company for the current year
    df['years_spent'] = year - df['date_employed'].dt.year

    for tenure in range(10, 31, 5):
        exact_years_df = df[df['years_spent'] == tenure]
        people = []
        for index, row in exact_years_df.iterrows():
            people.append({
                'Company': row['company'],
                'Full Name': row['full_name']
            })

        results.append({
            'Year': year,
            'Tenure': f'{tenure} years',
            'Number of People': len(people)
        })

# Create a DataFrame from the results
result_df = pd.DataFrame(results)

# Save the DataFrame to an Excel file
result_df.to_excel('employment_results.xlsx', index=False)
```

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('awards.csv')

# Convert the 'date_employed' column to datetime using 'month/day/year' format
df['date_employed'] = pd.to_datetime(df['date_employed'], format='%m/%d/%Y')

# Define the years you want to check
target_years = [2024, 2025, 2026]

# Create a result Excel writer object to save multiple DataFrames into a single Excel file
with pd.ExcelWriter('employment_results.xlsx', engine='xlsxwriter') as writer:
    for target_year in target_years:
        # Calculate the number of years spent in the company for the current target year
        df['years_spent'] = target_year - df['date_employed'].dt.year

        # Create a result DataFrame for the current target year
        result_df = pd.DataFrame(columns=['Company', 'Full Name', 'Date Employed'])

        # Loop through the target years and collect the names of people who have spent exactly those years by the current target year
        for years in target_years:
            exact_years_df = df[df['years_spent'] == years]
            people = []
            for index, row in exact_years_df.iterrows():
                people.append({
                    'Company': row['company'],
                    'Full Name': row['full_name'],
                    'Date Employed': row['date_employed']
                })

            # Append the data to the result DataFrame
            result_df = result_df.append(people, ignore_index=True)

        # Save the current result DataFrame to a separate sheet with the target year as the sheet name
        sheet_name = str(target_year)
        result_df.to_excel(writer, sheet_name=sheet_name, index=False)

print("Data saved to employment_results.xlsx")
```
